```python
#Valor
Refrigerante = 15
preco = 0
total = 0
refrigerantes = 0

#tabela preço
print("Bem-vindo a Pizzaria do Bruno Kostiuk")
print("-----------------------------------Cardápio-----------------------------------")

print("| Tamanho | Pizza Salgada (PS) | Pizza Doce (PD) |")
print("|---------|---------------------|------------------|")
print("| P       | R$ 30.00            | R$ 34.00         |")
print("| M       | R$ 45.00            | R$ 48.00         |")
print("| G       | R$ 60.00            | R$ 66.00         |")
print("")
print("| Refrigerantes 2L | R$ 15.00 |")
print("")
print("NÃO REALIZAMOS ENTREGAS PARA DISTÂNCIA ACIMA DE 25KM")
print("")

#pedido
while True:
    item = input("Entre com o item que deseja (Pizza/Refrigerante): ").lower()
    if item == "pizza":
        sabor = input("Entre com o sabor desejado (PS/PD): ").upper()
        tamanho = input("Entre com o tamanho desejado (P/M/G): ").upper()

        if (sabor == "PS"):
            if (tamanho == "P"):
                preco += 30
            elif (tamanho == "M"):
                preco += 45
            elif (tamanho == "G"):
                preco += 60
            else:
                print("TAMANHO INVALIDO")
                continue

        elif (sabor == "PD"):
            if (tamanho == "P"):
                preco += 34
            elif (tamanho == "M"):
                preco += 48
            elif (tamanho == "G"):
                preco += 66
            else:
                print("TAMANHO INVALIDO")
                continue
        else:
            print("SABOR INVALIDO")
            continue

#calculo

        total += preco
        print(f"Você pediu uma Pizza {sabor} no tamanho {tamanho}: R$ {preco:.2f}")

    elif item == "refrigerante":
        if refrigerantes < 2:
            preco = 15
            total += preco
            refrigerantes += 1
            print(f"Você pediu um Refrigerante de 2L: R$ {preco:.2f}")
        else:
            print("Quantidade máxima de Bebidas atingida")

    else:
        print("OU PIZZA OU REFRI")
        continue

    resposta = input("Deseja mais alguma coisa? (S/N): ").lower()
    if resposta != "s":
        break

#entrega

while True:
    distancia = float(input("Entre com a distância para entrega (apenas números em Km): "))
    if distancia <= 25:
        if distancia <= 5:
            custo_entrega = 0
        elif distancia < 10:
            custo_entrega = 10
        else:
            custo_entrega = 10 + (distancia - 10) * 1.10
        print(f"Entrega: R$ {custo_entrega:.2f}")
        total += custo_entrega
        break
    else:
        print("NÃO REALIZAMOS ENTREGAS PARA DISTÂNCIA ACIMA DE 25KM")

#total

print(f"O valor total a ser pago: R$ {total:.2f}")
```
